```python
class Perceptron:
    def __init__(self):
        self.w = None
        self.b = None
    
    def model(self,x):
        return 1 if (np.dot(self.w,x) >= self.b) else 0
        
        
        
        
    def predict(self,X):
        Y=[]
        for x in X:
            result = self.model(x)
            Y.append(result)
        return Y    
        
        
        
    def fit(self,X,Y):
        
        
         
        self.w = np.ones(X.shape[1])
        self.b = 0
        
        for x,y in zip(X,Y):
            y_pred = self.model(x)
            if y == 1 and y_pred == 0:
                self.w = self.w + x
                self.b = self.b -1
            
            elif y == 0 and y_pred == 1:
                self.w = self.w - x
                self.b = self.b +1
            
               
                
            
            
            
```

# CLASS SIGMOID NEURON 

```python
class SigmoidNeuron:
    
    def __init__(self):
        self.w = None
        self.b = None
    
    
    def perceptron(self,x):
        return np.dot(x,self.w.T)+ self.b
    
    
    
    def sigmoid(self,x):
        return 1/(1 + np.exp(-x))
    
    
    def grad_w(self, x, y):
        y_pred = self.sigmoid(self.perceptron(x))
        return (y_pred-y) * y_pred * (1 - y_pred) * x
    
    
    def grad_b(self, x, y):
        y_pred = self.sigmoid(self.perceptron(x))
        return (y_pred - y) * (y_pred) * (1 - y_pred)
    
    
    def fit(self, X, Y, epochs=1, learning_rate=1, initialize= True):
        
        #initialize w,b
        
        if initialize:
            self.w = np.random.randn(1, X.shape[1])
            self.b = 0
            
        for i in range(epochs):
            dw =0
            db =0
            for x,y in zip(X,Y):
                dw += self.grad_w(x,y)
                db += self.grad_b(x,y)
            self.w -= learning_rate * dw
            self.b -= learning_rate * db
                

    
```

```python
X = np.asarray([[2.5,2.5],[4,-1], [1,-4],[3,1.25],[2,4],[1,5]])
Y = [1,1,1,0,0,0]
```

```python
sn = SigmoidNeuron()
sn.fit(X,Y,1,0.25,True)
```

```python
sn.fit(X,Y,1,0.5,True)
for i in range(20):
    print(sn.w,sn.b)
    sn.fit(X,Y,1,0.5,False)
     
```
